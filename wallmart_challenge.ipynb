```python
df2=dat1.pivot_table('ScanCount', ['TripType', 'VisitNumber','Weekday'],['DepartmentDescription'], aggfunc=lambda x:len(x.unique()))
```

```python
df2.reset_index(inplace=True)
```

```python
#df2.head()
```

```python
dummies4= pd.get_dummies(df2['Weekday'])
df2=pd.concat([df2, dummies4], axis=1)
#df2.head()
```

```python
map_triptype= {3:0, 4:1, 5:2, 6:3, 7:4, 8:5, 9:6, 12:7, 14:8, 15:9, 18:10, 19:11, 20:12, 21:13, 22:14, 23:15,
               24:16, 25:17, 26:18, 27:19, 28:20, 29:21, 30:22, 31:23, 32:24, 33:25, 34:26, 35:27, 
               36:28, 37:29, 38:30, 39:31, 40:32, 41:33, 42:34, 43:35, 44:36, 999:37}
df2['TripType']=df2['TripType'].map(map_triptype)
#df2.head()
```

```python
df2.fillna(0, inplace=True)
#df2.head()
```

```python
#X= train_data.drop(train_data[[0,1,2,3,4,5,6]], axis=1)
X1=df2.drop(df2[[0,1,2]], axis=1)
```

```python
#Y = train_data['TripType']
Y1= df2['TripType']
```

```python
from sklearn.cross_validation import train_test_split
#X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.3)
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1, Y1, test_size= 0.25)
```

```python
# Using XGboost for classification

#dtrain = xgb.DMatrix(X_train, label=Y_train)
#dtest = xgb.DMatrix(X_test, label=Y_test)
dtrain = xgb.DMatrix(X1_train, label=Y1_train)
dtest = xgb.DMatrix(X1_test, label=Y1_test)
```

```python
# Tree Booster parameters

param = {'bst:max_depth':10, 'bst:eta':0.04, 'silent':1, 'objective':'multi:softmax'}
# max_depth: maximum depth of the tree
# eta: step size/shrinkage; After each boosting step, we can directly get the weights of new features and 
# eta actually shrinks the feature weights to make the boosting process more conservative.

# depth: 6, eta:0.03, num_round=1500 --> have the best results uptill now!!, 0.989

param['nthread'] = 1
# nthread: number of parallel threads used to run xgboost.

param['num_class'] = 38 # number of classes in the model

param['eval_metric'] = 'mlogloss' #['mlogloss', 'merror'] 
# “mlogloss”: Multiclass logloss
# “merror”: Multiclass classification error rate. It is calculated as #(wrong cases)/#(all cases).
# multi class log loss- penalizes false confident predictions 

#param['gamma']= 0.1  # minimum loss reduction required to make a further partition; 
# this value could be increased to 0.5

#param['subsample']= 0.75 
# this is can used to prevent overfitting, as XGBoost will randomly select 75% of the training data.

#param['colsample_bytree']= 0.75
# this is used to prevent overfitting; it will randomly select 75% of the columns in training data
# Specify validations set to watch performance
evallist  = [(dtest,'eval'), (dtrain,'train')]
```

```python
num_round = 500 # number of rounds of boosting
bst = xgb.train(param, dtrain, num_round, evallist)
```

```python
bst.save_model('model2')
```
