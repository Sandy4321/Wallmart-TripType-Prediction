```python
import pandas as pd
import numpy as np
import xgboost as xgb
```

```python
test_data1= pd.read_csv('/home/prateek/UVa/own work/kag1/wallmart recruiting challenge/test.csv')
test_data1.shape
```

```python
df3=test_data1.pivot_table('ScanCount', ['VisitNumber', 'Weekday'], ['DepartmentDescription'], aggfunc=np.sum)
df3.head()
```

```python
df3.reset_index(inplace=True)
df3.head()
```

```python
df3.fillna(0, inplace=True)
df3.head()
```

```python
dummies5= pd.get_dummies(df3['Weekday'])
df3=pd.concat([df3, dummies5], axis=1)
df3.head()
```

```python
X3=df3.drop(df3[[0,1]], axis=1)
X3.head()
```

```python
# loading the model
bst = xgb.Booster({'nthread':1}) #init model
bst.load_model("model1") # load data
```

```python
# predicting on the test data
dtest = xgb.DMatrix(X3)
ypred = bst.predict(dtest)
```

```python
# adding the VisitNumber to predictions
ypred_new=pd.DataFrame(ypred)

ypred_new['VisitNumber']= df3['VisitNumber']
```

```python
# renaming the column names
ypred_new.columns=['TripType','VisitNumber']
```

```python
ypred_new.shape
```

```python
# There is a difference in dimensions of the original test data and resultant dataframe , mainly because of the null
# values present in the test data,
```

```python
# unique visit number is the original dataset
len(test_data1['VisitNumber'].unique())
```

```python
# find the difference in length of the test dataframe and the result dataframe and sorting them
vnum=list(set(test_data1.VisitNumber)-set(df3.VisitNumber))
sorted(vnum)
```

```python
# finding the dataframe with left out visitnumbers
leftout_visit=test_data1.loc[test_data1['VisitNumber'].isin(vnum)] # for single value use this '=='; and for multiple
# values use 'isin'
# finding the unique set
visitname=pd.DataFrame(leftout_visit['VisitNumber'].unique())
```

```python
# adding a new column
visitname['TripType']=37
```

```python
# renaming the columns
visitname.columns=['VisitNumber', 'TripType']
visitname.shape
```

```python
# adding to the resultant dataframe
ypred_new2=pd.concat([ypred_new, visitname], axis=0)
ypred_new2.head()
```

```python
# checking the new dimensions
ypred_new2.shape
```

```python
# the old dimensions
ypred_new.shape
```

```python
# converting 'TripType' to dummy variables
dummies6= pd.get_dummies(ypred_new2['TripType'])
data_kag=pd.concat([ypred_new2, dummies6], axis=1)
data_kag.fillna(0, inplace=True)
data_kag.shape
```

```python
data_kag.columns.values
```

```python
# dropping the 'TripType' Column and adding the appropriate column names
data_kag2=data_kag.drop(data_kag[['TripType']], axis=1)
data_kag2.columns=['VisitNumber', 'TripType_3', 'TripType_4', 'TripType_5', 'TripType_6', 'TripType_7', 'TripType_8',
                  'TripType_9', 'TripType_12', 'TripType_14', 'TripType_15', 'TripType_18', 'TripType_19',
                  'TripType_20', 'TripType_21', 'TripType_22', 'TripType_23', 'TripType_24', 'TripType_25',
                  'TripType_26', 'TripType_27', 'TripType_28', 'TripType_29', 'TripType_30', 'TripType_31',
                  'TripType_32', 'TripType_33', 'TripType_34', 'TripType_35', 'TripType_36', 'TripType_37',
                  'TripType_38', 'TripType_39', 'TripType_40', 'TripType_41', 'TripType_42', 'TripType_43',
                  'TripType_44', 'TripType_999']
#data_kag2['TripType_14']=0
data_kag2.head()
```
